#When considering 'cost per impression', what day of the week works best? 
#(What day is it cheapest to generate impressions) 
#What day works worst? (What day is most expensive)

```python
import pandas as pd
df = pd.read_csv('Travel Pony Facebook.csv')
print(df)
```

```python
list(df)
```

```python
df_spend = df["Amount Spent (USD)"]
print(df_spend.head())
```

```python
df_imp = df["Impressions"]
print(df_imp.head())
```

```python
df_costperimp = df_spend/df_imp
print(df_costperimp.head())
```

```python
df.insert(loc=0,column="cost per imp", value=df_costperimp)
print(list(df))
```

```python
import arrow
week = []
for i in range(len(df)):
    dt = arrow.get(df['Start Date'].iloc[i], 'M/D/YY')
    dayoftheweek = dt.isoweekday()
    week.append(dayoftheweek)
len(week)
print(len(week))
```

```python
df.insert(loc=0, column='day of the week', value=week)
print(df)
```

```python
temp = df.groupby(['day of the week'])['cost per imp'].mean()

print(temp)
```

#3 What variable most strongly predicts unique clicks? 
#What does that mean practically? (respond in a tweet or less)

```python
import statsmodels.api as st

Y = df['Unique Clicks']
X = df[['Reach', 'Frequency']]

m = st.OLS(Y, X).fit()
predict = m.predict(X)

print(m.summary())
```
